```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('nfl_offensive_stats.csv')
```

```python
df.head()
```

```python
df = df.set_index('player')
df.head()
```

```python
aggregated_df = df.groupby('player').sum()
aggregated_df.head()
```

```python
cols_drop = [ 'offense','off_pct', 'vis_score', 'home_score', 'OT', 'Temperature', 'Humidity',
            'Wind_Speed', 'Vegas_Line', 'Over_Under','Total_DKP', 'Off_DKP', 'Total_FDP', 'Off_FDP',
            'Total_SDP', 'Off_SDP']
compare_df = aggregated_df.drop(cols_drop,axis=1)

compare_df.reset_index(inplace=True)
compare_df[300:320]
```

```python
compare_df.columns
```

```python
columns_compare = ['pass_cmp', 'pass_att', 'pass_yds', 'pass_td', 'pass_int',
       'pass_sacked', 'pass_sacked_yds', 'pass_long', 'pass_rating',
       'rush_att', 'rush_yds', 'rush_td', 'rush_long', 'targets', 'rec',
       'rec_yds', 'rec_td', 'rec_long', 'fumbles_lost', 'rush_scrambles',
       'designed_rush_att', 'comb_pass_rush_play', 'comb_pass_play',
       'comb_rush_play','pass_target_yds', 'pass_poor_throws',
       'rush_yds_before_contact', 'rush_yac',
       'rush_broken_tackles', 'rec_air_yds', 'rec_yac', 'rec_drops']
```

```python
players_to_compare = ['Lamar Jackson','Tyreek Hill','Gus Edwards','Mark Andrews','Christian McCaffrey']
```

```python
selected_data = compare_df[columns_compare]
```

```python
cosine_similarity_df = cosine_similarity(selected_data)
```

```python
similarity_df = pd.DataFrame(cosine_similarity_df,columns = compare_df['player'],index = compare_df['player'])
similarity_df.head()
```

```python
top_similar_dict = {}

for player in players_to_compare:
    similarity_score = similarity_df.loc[player]
    
    similarity_sorted = similarity_score.sort_values(ascending=False)
    
    top_10 = similarity_sorted[1:11]
    top_similar_dict[player] = top_10
```

```python
for player, sim_player in top_similar_dict.items():
    print(f'Top 10 similar players for {player}:')
    print(sim_player)
    print('\n')
    
```
